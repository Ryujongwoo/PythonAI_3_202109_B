numpy의 sum() 함수의 작동 방식

```python
# sum() 함수의 axis 속성의 기본값은 'None'이고 axis는 대상 데이터의 차원 보다 작은 값을 설정한다.
# 위의 예제 데이터의 차원이 3이기 때문에 axis 속성은 2까지 지정할 수 있다.
print(v.ndim) # ndim: numpy 배열의 차원(Dimension)을 얻어온다.
```

```python
# axis=None
# axis를 기본값으로 실행하면 대상 데이터의 모든 요소의 합계를 계산해 리턴한다.
# 대상 데이터에 포함된 모든 요소를 단순하게 합산하는 연산이 실행된다.
print(v.shape)
print(v.sum())
print(v.sum(axis=None))
```

```python
# axis=0
# axis=0는 x축을 기준으로 계산한 합계를 리턴한다.
# shape이 (4, 2, 4)인 데이터의 x축을 기준으로 합계를 계산하고 결과 shqpe은 면이 제거되고 (2, 4)이 리턴된다.
result = v.sum(axis=0)
print(result)
print(result.shape)
print(result.ndim)
```

```python
# axis=0은 가장 외곽의 괄호를 제거하는 이미지를 생각하면 된다.
# 아래 그림처름 가장 외곽의 괄호를 제거하고 각 행의 데이터는 위치별로 합계를 계산한다.
Image('./numpyImage/shape3.png', width='800')
```
